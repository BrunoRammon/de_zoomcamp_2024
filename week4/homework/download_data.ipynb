```python
import pandas as pd
import pyarrow.parquet as pq
import pyarrow as pa
import requests
import os  
```

```python

years = {
    'green': [2019, 2020],
    'fhv': [2019],
    'yellow': [2019, 2020],
}
for type_taxi in years.keys():
    dfs = []
    for year in years[type_taxi]:
        for month in range(1,13):
            file_name = f"{type_taxi}_tripdata_{year}-{month:02d}.csv.gz"
            url = f"https://github.com/DataTalksClub/nyc-tlc-data/releases/download/{type_taxi}/{file_name}"
            local_folder = f'./ny_taxi_data/{type_taxi}'
            os.makedirs(local_folder, exist_ok=True)
            response = requests.get(url)
            if response.status_code == 200:
                # Write the content to the file
                with open(f'{local_folder}/{file_name}', 'wb') as file:
                    file.write(response.content)
                print(f"File downloaded to: {local_folder}")
            else:
                print(f"Failed to download file. Status code: {response.status_code}")
```

```python
dfs = []
for month in range(1,13):
    url = f'https://d37ci6vzurychx.cloudfront.net/trip-data/fhv_tripdata_2019-{month:02d}.parquet'
    print(url)
    try:
        schema = pa.schema([
            ('dispatching_base_num', pa.string()),
            ('pickup_datetime', pa.string()),
            ('dropOff_datetime', pa.string()),
            ('PUlocationID', pa.float64()),
            ('DOlocationID', pa.float64()),
            ('SR_Flag', pa.float64()),
            ('Affiliated_base_number', pa.string())
        ])
        df = pd.read_parquet(url, schema=schema)
    except pa.ArrowInvalid:
        response = requests.get(url)
        if response.status_code == 200:
            # Write the content to the file
            with open(f'./{file_name}', 'wb') as file:
                file.write(response.content)
        df = pq.read_table(f'./{file_name}', schema=schema)
        df = df.to_pandas()
        print(f'Month {month:02d} not included')
    
    dfs.append(df)

df = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
month=9
file_name = f'fhv_tripdata_2019-{month:02d}.parquet'
url = f'https://d37ci6vzurychx.cloudfront.net/trip-data/{file_name}'

response = requests.get(url)
if response.status_code == 200:
    # Write the content to the file
    with open(f'./{file_name}', 'wb') as file:
        file.write(response.content)

schema = pa.schema([
    ('dispatching_base_num', pa.string()),
    ('pickup_datetime', pa.string()),
    ('dropOff_datetime', pa.string()),
    ('PUlocationID', pa.float64()),
    ('DOlocationID', pa.float64()),
    ('SR_Flag', pa.float64()),
    ('Affiliated_base_number', pa.string())
])
df = pq.read_table(f'./{file_name}', schema=schema)

df.to_pandas().assign(
    pickup_datetime = lambda df: pd.to_datetime(df.pickup_datetime.str.replace('3019','2019').str.replace('2919','2019').str.replace('5201','2019')),
    dropOff_datetime = lambda df: pd.to_datetime(df.dropOff_datetime.str.replace('3019','2019').str.replace('2919','2019').str.replace('5201','2019')),
)
```
